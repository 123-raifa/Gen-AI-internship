## Problem 1: Employee Performance Bonus Eligibility

```python
# Problem 1: Employee Performance Bonus Eligibility
# Identify the highest score and all employees who achieved it

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score among all employees
top_score = max(employees.values())

# Collect all employees who achieved the top score (handles ties)
top_performers = [name for name, score in employees.items() if score == top_score]

# Display the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {top_score})")
```

## Problem 2: Search Query Keyword Analysis

```python
# Problem 2: Search Query Keyword Analysis
# Count keyword frequency and display words that appear more than once

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase to treat 'Buy' and 'buy' as the same keyword
query_lower = query.lower()

# Remove common punctuation from the query
query_clean = query_lower.translate(str.maketrans("", "", string.punctuation))

# Split into individual keywords
keywords = query_clean.split()

# Count frequency of each keyword using a dictionary
keyword_count = {}
for word in keywords:
    keyword_count[word] = keyword_count.get(word, 0) + 1

# Filter only keywords that appear more than once
frequent_keywords = {word: count for word, count in keyword_count.items() if count > 1}

# Display the result
print(frequent_keywords)
```

## Problem 3: Sensor Data Validation

```python
# Problem 3: Sensor Data Validation
# Identify even-numbered readings (valid) and store as (hour_index, value) tuples

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Filter valid readings: even values only, stored as (index, value) tuples
# Index represents the hour of the reading
valid_readings = [
    (index, value)
    for index, value in enumerate(sensor_readings)
    if value % 2 == 0  # Even check — odd readings are invalid
]

# Display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

```python
# Problem 4: Email Domain Usage Analysis
# Count users per domain and calculate percentage usage

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

total_users = len(emails)

# Extract domain from each email and count occurrences
domain_count = {}
for email in emails:
    domain = email.split("@")[1]  # Everything after '@' is the domain
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Display domain usage with percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

## Problem 5: Sales Spike Detection

```python
# Problem 5: Sales Spike Detection
# Find days where sales exceed 30% above the average

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Threshold: 30% above average
spike_threshold = average_sales * 1.30

# Identify spike days (day numbers are 1-based)
print("Sales Spike Days (>30% above average):")
for day_index, sale in enumerate(sales):
    if sale > spike_threshold:
        print(f"Day {day_index + 1}: {sale}")
```

## Problem 6: Duplicate User ID Detection

```python
# Problem 6: Duplicate User ID Detection
# Find user IDs that appear more than once and show their count

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences of each user ID
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Filter and display only duplicate IDs (appearing more than once)
print("Duplicate User IDs:")
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
