## Problem 1: Smart Parking Lot Management System

```python
# Problem 1: Smart Parking Lot Management System
# Tracks vehicle entry/exit, counts parked vehicles, and alerts if capacity exceeded

def parking_lot_manager(capacity, vehicle_logs):
    current_count = 0  # Tracks number of vehicles currently parked
    peak_usage = 0     # Tracks the highest number of vehicles at any point

    for log in vehicle_logs:
        if log == "IN":
            current_count += 1  # Vehicle entered
        elif log == "OUT" and current_count > 0:
            current_count -= 1  # Vehicle exited (cannot go below 0)

        # Update peak usage if current count is the highest seen so far
        if current_count > peak_usage:
            peak_usage = current_count

    # Determine parking status based on capacity
    if current_count >= capacity:
        status = "Full - No Space Available!"
    elif current_count >= capacity * 0.8:
        status = "Almost Full"
    else:
        status = "Available"

    # Display results
    print(f"Currently Parked Vehicles: {current_count}")
    print(f"Peak Parking Usage: {peak_usage} vehicles")
    print(f"Parking Status: {status}")

    # Alert if parking exceeded capacity at any point
    if peak_usage > capacity:
        print(f"ALERT: Parking exceeded capacity! Peak was {peak_usage} vs Capacity {capacity}")


# --- Sample Input ---
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_lot_manager(parking_capacity, vehicle_logs)
```

## Problem 2: Online Food Delivery Time Estimator

```python
# Problem 2: Online Food Delivery Time Estimator
# Estimates delivery time based on distance, traffic level, and weather condition

def estimate_delivery_time(distance_km, traffic_level, weather):
    # Base speed: 40 km/h under normal conditions
    base_speed = 40

    # Calculate base delivery time in minutes
    base_time = (distance_km / base_speed) * 60

    # Add delay based on traffic level
    if traffic_level.lower() == "low":
        traffic_delay = 0       # No extra delay
    elif traffic_level.lower() == "medium":
        traffic_delay = 10      # 10 minutes added
    elif traffic_level.lower() == "high":
        traffic_delay = 20      # 20 minutes added
    else:
        traffic_delay = 0

    # Add delay based on weather condition
    if weather.lower() == "sunny":
        weather_delay = 0       # No extra delay
    elif weather.lower() == "cloudy":
        weather_delay = 5       # 5 minutes added
    elif weather.lower() == "rainy":
        weather_delay = 15      # 15 minutes added
    else:
        weather_delay = 0

    # Total estimated delivery time
    total_time = round(base_time + traffic_delay + weather_delay)

    print(f"Estimated Delivery Time: {total_time} minutes")


# --- Sample Input ---
estimate_delivery_time(distance_km=8, traffic_level="High", weather="Rainy")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
# Calculates occupancy percentage, determines show status, and suggests extra shows

def analyze_theatre_occupancy(total_seats, booked_seats):
    # Count total booked seats (sum of all 1s in the list)
    total_booked = sum(booked_seats)

    # Calculate occupancy percentage
    occupancy_percentage = (total_booked / total_seats) * 100

    # Determine show status based on occupancy
    if occupancy_percentage == 100:
        status = "Housefull"
        suggestion = "Open additional shows immediately!"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
        suggestion = "Consider opening an additional show."
    elif occupancy_percentage >= 50:
        status = "Filling Up"
        suggestion = "No additional shows needed yet."
    else:
        status = "Available"
        suggestion = "No additional shows needed."

    # Display results
    print(f"Occupancy: {int(occupancy_percentage)}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")


# --- Sample Input ---
total_seats = 200
booked_seats = [1] * 150  # 150 booked seats represented as list of 1s

analyze_theatre_occupancy(total_seats, booked_seats)
```

## Problem 4: Cloud Server Load Classification System

```python
# Problem 4: Cloud Server Load Classification System
# Classifies server load based on average CPU usage readings

def classify_server_load(cpu_readings):
    # Calculate average CPU usage across all readings
    average_cpu = sum(cpu_readings) / len(cpu_readings)

    # Classify server status based on average CPU load
    if average_cpu < 50:
        status = "Normal"    # Low load, everything fine
    elif average_cpu <= 80:
        status = "Warning"   # Moderate load, monitor closely
    else:
        status = "Critical"  # High load, immediate action needed

    # Display results
    print(f"Average CPU Load: {int(average_cpu)}%")
    print(f"Server Status: {status}")


# --- Sample Input ---
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

## Problem 5: Smart Classroom Resource Usage Monitor

```python
# Problem 5: Smart Classroom Resource Usage Monitor
# Identifies overused classroom resources and triggers energy alerts

def monitor_classroom_resources(resource_usage):
    # Threshold: resources used more than 8 hours are considered overused
    overuse_threshold = 8

    overused = []  # List to store names of overused resources

    # Check each resource against the threshold
    for resource, hours in resource_usage.items():
        if hours > overuse_threshold:
            overused.append(resource)

    # Trigger energy alert if any resource is overused
    energy_alert = "Yes" if overused else "No"

    # Display results
    if overused:
        print(f"Overused Resources: {', '.join(overused)}")
    else:
        print("Overused Resources: None")

    print(f"Energy Alert: {energy_alert}")


# --- Sample Input ---
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_classroom_resources(resource_usage)
```

## Problem 6: Online Event Registration Capacity Controller

```python
# Problem 6: Online Event Registration Capacity Controller
# Manages registrations, prevents overbooking, and triggers waitlist mode

def manage_event_registration(event_capacity, total_registrations):
    # Confirmed registrations cannot exceed event capacity
    if total_registrations <= event_capacity:
        confirmed = total_registrations
        waitlisted = 0
        reg_status = "Open"      # Still space available
    else:
        confirmed = event_capacity                        # Cap at capacity
        waitlisted = total_registrations - event_capacity # Overflow goes to waitlist
        reg_status = "Closed"    # No more registrations accepted

    # Display results
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {reg_status}")


# --- Sample Input ---
event_capacity = 100
registrations = 105

manage_event_registration(event_capacity, registrations)
```
