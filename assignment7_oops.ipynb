## Problem 1: Smart Light Controller

```python
# Problem 1: Smart Light Controller
# A class to control a smart light — turn it ON/OFF and display its status

class SmartLight:
    def __init__(self, light_name):
        # Initialize light with a name and default status as OFF
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        # Turn the light ON
        self.status = "ON"

    def turn_off(self):
        # Turn the light OFF
        self.status = "OFF"

    def display_status(self):
        # Display the current status of the light
        print(f"{self.light_name} is {self.status}")


# --- Sample Input ---
light = SmartLight("Bedroom Light")
light.turn_on()          # Set action to ON
light.display_status()   # Display status
```

## Problem 2: Employee ID Card System

```python
# Problem 2: Employee ID Card System
# A class to store employee details and display a formatted ID card

class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        # Store employee name, ID, and department
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        # Print formatted employee ID card
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# --- Sample Input ---
emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display_card()
```

## Problem 3: Mobile Contact Record

```python
# Problem 3: Mobile Contact Record
# A class to store and display a contact's name and phone number

class MobileContact:
    def __init__(self, contact_name, phone_number):
        # Store contact name and phone number
        self.contact_name = contact_name
        self.phone_number = phone_number

    def display_contact(self):
        # Display saved contact details
        print("Contact Saved")
        print(f"Name: {self.contact_name}")
        print(f"Phone: {self.phone_number}")


# --- Sample Input ---
contact = MobileContact("Anita", "9876543210")
contact.display_contact()
```

## Problem 4: Product Price Tag Generator

```python
# Problem 4: Product Price Tag Generator
# A class to store product details and print a formatted price tag

class ProductPriceTag:
    def __init__(self, product_name, price):
        # Store product name and price
        self.product_name = product_name
        self.price = price

    def display_price_tag(self):
        # Print a formatted price tag with rupee symbol
        print(f"Product: {self.product_name}")
        print(f"Price: \u20b9{self.price}")  # \u20b9 is the ₹ symbol


# --- Sample Input ---
product = ProductPriceTag("Headphones", 2499)
product.display_price_tag()
```

## Problem 5: Movie Rating Display System

```python
# Problem 5: Movie Rating Display System
# A class to store movie name and rating and display details

class MovieRating:
    def __init__(self, movie_name, rating):
        # Store movie name and rating out of 5
        self.movie_name = movie_name
        self.rating = rating

    def display_movie(self):
        # Display movie name and its rating
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")


# --- Sample Input ---
movie = MovieRating("Inception", 4.8)
movie.display_movie()
```

## Problem 6: Delivery Address Manager

```python
# Problem 6: Delivery Address Manager
# A class to store customer name and delivery address and print delivery details

class DeliveryAddress:
    def __init__(self, customer_name, address):
        # Store customer name and delivery address
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        # Print formatted delivery information
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# --- Sample Input ---
delivery = DeliveryAddress("Suman", "Hyderabad")
delivery.display_delivery_details()
```
